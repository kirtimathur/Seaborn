```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

### univariate analysis of numeric: boxplot

```python
df=sns.load_dataset("tips")
df.head()
```

```python
sns.boxplot(df["total_bill"]);
```

```python
sns.catplot(x="day",y="tip",data=df,kind="box"); #categoric-numeric
```

```python
sns.catplot(x="day",y="tip",data=df,kind="box",hue="sex");
```

```python
sns.catplot(x="day",y="tip",data=df,kind="box",hue="sex",dodge=False);
```

### Violin plot: to check density and outliers

## correlation: 
- to check realtion b/w two variables
- value is (-1,1)
- if corr=0, both variables are independent, no realtion b/w variables
- if corr>0, then +ve correlation (directly proportional, increase or decrease together)
    - close to 1 means highly correlated
- if corr<0, then -ve correlation
    - close to -1 low correlated

```python
df.corr() #corr with itself is one
#tip and total_bill are moderately linear
#comparing to size with tip and bill, bill is more related
```

```python
sns.heatmap(df.corr()); #lighter means high corr
```

```python
sns.heatmap(df.corr(),annot=True); #annot to display the corr no.s
```

```python
#to check for specific cols
df[["total_bill","tip"]].corr() #to find the factor due to which sale is increasing
```
