#### Countplot: 
- used for univariate analysis
- used for categorical data
- same as value counts (frequency of unique element in the data)

```python
df=sns.load_dataset("tips")
df.head()
```

#### Count of males and females visiting the restaurant (in pandas use value_counts, here countplot)

```python
sns.countplot(x="sex",data=df); #univariate anlysis of all categorical cols
```

#### Number of smokers and non- smoker visiting the restaurant 

```python
sns.countplot(x="smoker",data=df);
```

#### Peak time customer visited the restaurant

```python
sns.countplot(x="time",data=df,order=["Dinner","Lunch"]); #order=[] to change the order
```

#### The peak day

```python
sns.countplot(x="day",data=df);
```

#### Peak day for females and males 
- (both categorical columns, bivariate analysis, count plot with hue)

```python
sns.countplot(x="day",hue="sex",data=df);
```

- Females visited more on Thursday and males visted more on Saturday and Sunday.

#### smokers and non smokers who were males and females

```python
sns.countplot(x="sex",hue="smoker",data=df);
```

- Both females and males had more non smokers
