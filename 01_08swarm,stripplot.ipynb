```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Swarmplot: for specific scatters

```python
df=sns.load_dataset("tips")
df.head()
```

```python
sns.catplot(x="day",y="total_bill",data=df,kind="swarm",estimator=sum); 
```

- In total, frequency of people and their bill
- Mostly bill given on sat

```python
sns.catplot(x="time",y="tip",data=df,kind="swarm");
```

```python
#in strip plot can't count how much bill was given, like in swarm can count who gave 17 
sns.catplot(x="day",y="total_bill",data=df); 
```
