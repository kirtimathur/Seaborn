```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

- seaborn is more advance & highly interactive as compared to matplotlib
    - allows to multivariate analysis
    - gives statistical analysis in 1 frame

## numerical plots
- related to numeric data (continous and discrete)
- plots:
    - relplot()
    - scatterplot()
    - lineplot()

### scatterplot : used for bivariate analysis and both columns are numeric (continous type)
- plot using either relplot() or scatterplot()
    - scatterplot gives scatter plot by default
    - but relplot is genric(can make other plots) : relplot means relational plot

```python
df=sns.load_dataset("tips")
df.head()
```

```python
#trend of total_bill and tip
sns.relplot(x="total_bill",y="tip",data=df);
```

- insight: a sort of linear relationship: if bill increases the tips sort of tends to increase

```python
sns.relplot(x="total_bill",y="tip",hue="sex",data=df); #multivariate analysis, getting sex along with bills and tips
#in hue always give categorical type
```

- insights: lowest bill payer was a female and highest bill payer was a male
    - one female with 45 bill still gave less tip
    - most females lie in bill from 10 to 30 and tips  2 to 4

```python
sns.relplot(x="total_bill",y="tip",hue="smoker",data=df);
```

- Mostly non smokers lied in greater tip and total bill region

```python
sns.relplot(x="total_bill",y="tip",hue="sex",style="smoker",data=df); #two categories in one
```

```python
sns.relplot(x="total_bill",y="tip",hue="size",data=df); 
```

- it's not necessary with increasing in family size bill size increases
- light bubbles show ones with small family size have paid less tip

```python
sns.relplot(x="total_bill",y="tip",hue="time",data=df);
```

- most people came in for dinner
- people who came for lunch mostly paid lesser tips
- maximum bill payer and tip giver in dinner time

```python
sns.relplot(x="total_bill",y="tip",hue="sex",size="size",style="smoker",data=df,sizes=(50,200));
#hue,size,style takes categorical data
```

```python
sns.scatterplot(x="total_bill",y="tip",data=df); 
```
